```python
import pandas as pd
import matplotlib

# Load the CSV file
data = pd.read_csv('data/imdb.csv')
```

```python
# startYear column is an int
# lets print a bar chart of the number of movies per year
# first we need to group the data by year
data_by_year = data.groupby('startYear').size()

# eliminate data that isnt ints
data_by_year = data_by_year[data_by_year.index.str.isdigit()]

# then we can plot it, it's wide so we'll make it taller and chop off years with 10 or fewer movies
data_by_year.plot.bar( title='Number of movies per year', figsize=(20,10) )
```

```python
# let's do a histogram of the averageRating next to see how the ratings are distributed
data['averageRating'].hist( bins=100, figsize=(20,10) )
```

```python
# now finally, lets do a pie chart of the genres
# first we need to split the genres column into a list of genres
data['genres'] = data['genres'].str.split(',')

# then we need to explode the list so that each genre is its own row
data = data.explode('genres')

# let's ignote the \N values
data = data[data['genres'] != '\\N']

# then we can group by genre and count the number of movies in each genre
data_by_genre = data.groupby('genres').size()

# then we can plot it
data_by_genre.plot.pie( title='Number of movies per genre', figsize=(20,10) )
```
